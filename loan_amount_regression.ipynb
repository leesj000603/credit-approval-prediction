## 컬럼 설명

- no_of_dependents: 가족의 부양자 수. 정수 값으로 표현되며, 각 대출 신청자의 부양 가족 수를 나타냅니다.

- education: 대출 신청자의 교육 수준. 문자열 값으로 표현되며, 대출 신청자가 대졸, 중졸 등의 교육 수준을 나타냅니다.

- self_employed: 대출 신청자의 자영업 여부. 문자열 값으로 표현되며, 대출 신청자가 자영업인 경우를 나타냅니다.

- income_annum: 연간 수입. 정수 값으로 표현되며, 대출 신청자의 연간 총 수입을 나타냅니다.

- loan_amount: 대출 금액. 정수 값으로 표현되며, 대출 신청자가 신청한 대출의 금액을 나타냅니다.

- loan_term: 대출 상환 기간. 정수 값으로 표현되며, 대출 신청자가 선택한 대출의 상환 기간을 나타냅니다.

- cibil_score: CIBIL 신용 점수. 정수 값으로 표현되며, 대출 신청자의 신용 상태를 나타내는 지표 중 하나일 것입니다.

- residential_assets_value: 거주용 자산 가치. 정수 값으로 표현되며, 대출 신청자의 거주용 자산의 가치를 나타냅니다.

- commercial_assets_value: 상업용 자산 가치. 정수 값으로 표현되며, 대출 신청자의 상업용 자산의 가치를 나타냅니다.

- luxury_assets_value: 고급 자산 가치. 정수 값으로 표현되며, 대출 신청자의 고급 자산의 가치를 나타냅니다.

- bank_asset_value: 은행 자산 가치. 정수 값으로 표현되며, 대출 신청자의 은행 자산의 가치를 나타냅니다.

- loan_status: 대출 상태. 문자열 값으로 표현되며, 대출이 승인되었는지 여부를 나타냅니다 (예: 'approved' 또는 'rejected').

```python
from sklearn.metrics import mean_squared_error


my_predictions = {}
my_pred = None
my_actual = None
my_name = None

colors = ['r', 'c', 'm', 'y', 'k', 'khaki', 'teal', 'orchid', 'sandybrown',
          'greenyellow', 'dodgerblue', 'deepskyblue', 'rosybrown', 'firebrick',
          'deeppink', 'crimson', 'salmon', 'darkred', 'olivedrab', 'olive', 
          'forestgreen', 'royalblue', 'indigo', 'navy', 'mediumpurple', 'chocolate',
          'gold', 'darkorange', 'seagreen', 'turquoise', 'steelblue', 'slategray', 
          'peru', 'midnightblue', 'slateblue', 'dimgray', 'cadetblue', 'tomato'
         ]

def plot_predictions(name_, pred, actual):
    df = pd.DataFrame({'prediction': pred, 'actual': actual})
    df = df.sort_values(by='actual').reset_index(drop=True)

    plt.figure(figsize=(11, 8))
    plt.scatter(df.index, df['prediction'], marker='x', color='r')
    plt.scatter(df.index, df['actual'], alpha=0.7, marker='o', color='black')
    plt.title(name_, fontsize=15)
    plt.legend(['prediction', 'actual'], fontsize=12)
    plt.show()

def rmse_eval(name_, pred, actual):
    global my_predictions, colors, my_pred, my_actual, my_name
    
    my_name = name_
    my_pred = pred
    my_actual = actual
    
    plot_predictions(name_, pred, actual)
    
    mse = mean_squared_error(pred, actual)
    my_predictions[name_] = np.sqrt(mse)
    
    y_value = sorted(my_predictions.items(), key=lambda x: x[1], reverse=True)
    
    df = pd.DataFrame(y_value, columns=['model', 'rmse'])
    print(df)
    min_ = df['rmse'].min() - 10
    max_ = df['rmse'].max() + 10
    
    length = len(df) / 2
    
    plt.figure(figsize=(9, length))
    ax = plt.subplot()
    ax.set_yticks(np.arange(len(df)))
    ax.set_yticklabels(df['model'], fontsize=12)
    bars = ax.barh(np.arange(len(df)), df['rmse'], height=0.3)
    
    for i, v in enumerate(df['rmse']):
        idx = np.random.choice(len(colors))
        bars[i].set_color(colors[idx])
        ax.text(v + 2, i, str(round(v, 3)), color='k', fontsize=12, fontweight='bold', verticalalignment='center')
        
    plt.title('RMSE Error', fontsize=16)
    plt.xlim(min_, max_)
    
    plt.show()

def add_model(name_, pred, actual):
    global my_predictions, my_pred, my_actual, my_name
    my_name = name_
    my_pred = pred
    my_actual = actual
    
    mse = mean_squared_error(pred, actual)
    my_predictions[name_] = np.sqrt(mse)

def remove_model(name_):
    global my_predictions
    try:
        del my_predictions[name_]
    except KeyError:
        return False
    return True

def plot_all():
    global my_predictions, my_pred, my_actual, my_name
    
    plot_predictions(my_name, my_pred, my_actual)
    
    y_value = sorted(my_predictions.items(), key=lambda x: x[1], reverse=True)
    
    df = pd.DataFrame(y_value, columns=['model', 'rmse'])
    print(df)
    min_ = df['rmse'].min() - 10
    max_ = df['rmse'].max() + 10
    
    length = len(df) / 2
    
    plt.figure(figsize=(9, length))
    ax = plt.subplot()
    ax.set_yticks(np.arange(len(df)))
    ax.set_yticklabels(df['model'], fontsize=12)
    bars = ax.barh(np.arange(len(df)), df['rmse'], height=0.3)
    
    for i, v in enumerate(df['rmse']):
        idx = np.random.choice(len(colors))
        bars[i].set_color(colors[idx])
        ax.text(v + 2, i, str(round(v, 3)), color='k', fontsize=12, fontweight='bold', verticalalignment='center')
        
    plt.title('RMSE Error', fontsize=16)
    plt.xlim(min_, max_)
    
    plt.show()
    
def plot_coef(columns, coef):
    coef_df = pd.DataFrame(list(zip(columns, coef)))
    coef_df.columns=['feature', 'coef']
    coef_df = coef_df.sort_values('coef', ascending=False).reset_index(drop=True)

    fig, ax = plt.subplots(figsize=(9, 7))
    ax.barh(np.arange(len(coef_df)), coef_df['coef'])
    idx = np.arange(len(coef_df))
    ax.set_yticks(idx)
    ax.set_yticklabels(coef_df['feature'])
    fig.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet, Ridge, Lasso
from sklearn.metrics import r2_score
import numpy as np

# 데이터 및 모델을 적절히 설정
# X_train, X_test, y_train, y_test 설정 및 모델 초기화 등...

# 각 모델 학습 및 예측
models = {
    "ElasticNet": ElasticNet(alpha=0.001),
    "Ridge": Ridge(alpha=10),
    "Lasso": Lasso(alpha=10),
}

scores = {}

for name, model in models.items():
    model.fit(X_train_stand, y_train)
    predictions = model.predict(X_test_stand)
    score = r2_score(y_test, predictions)
    scores[name] = score

# 가로 막대 그래프로 시각화
model_names = list(scores.keys())
model_scores = list(scores.values())

fig, ax = plt.subplots()

bars = ax.barh(model_names, model_scores, color=['skyblue', 'lightgreen', 'lightcoral'])

# 각 바에 값 표시
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.2f}', 
             va='center', ha='left', color='black')

ax.set_xlabel("R^2 Score")
ax.set_title("Model Comparison")

plt.show()
```
